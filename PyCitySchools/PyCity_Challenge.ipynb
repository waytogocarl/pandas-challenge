### District Summary

```python
#set dependencies
import pandas as pd
from pathlib import Path
```

```python
#load files
load_school_data = Path(".", "schools_complete.csv")
load_student_data = Path(".", "students_complete.csv")
```

```python
#read files
school_data = pd.read_csv(load_school_data)
student_data = pd.read_csv(load_student_data)
```

```python
#combine the data
complete_school_data = pd.merge(student_data, school_data, how="left", on="school_name")
complete_school_data.head()
```

```python
#calculate the total number of unique schools
school_count = len(complete_school_data["school_name"].unique())
print(school_count)
```

```python
#calculate the total number of students
student_count = sum(complete_school_data["size"].unique())
#or complete_school_data{"Student ID"]count.()
print(student_count)
```

```python
#calculate the total budget
total_budget = sum(complete_school_data["budget"].unique())
print(total_budget)
```

```python
#calculate the average (mean) math score
average_math_score = complete_school_data["math_score"].mean()
print(average_math_score)
```

```python
#calculate the average (mean) reading score
average_reading_score = complete_school_data["reading_score"].mean()
print(average_reading_score)
```

```python
#calculate the percentage of students who passed math (math scores >= 70)
passing_math_count = complete_school_data[(complete_school_data["math_score"] >= 70)].count()["student_name"]
passing_math_percentage = passing_math_count / float(student_count) * 100
print(passing_math_percentage)
```

```python
#calculate the percentage of students who passed reading (hint: look at how the math percentage was calculated)
passing_reading_count = complete_school_data[(complete_school_data["reading_score"] >= 70)].count()["student_name"]
passing_reading_percentage = passing_reading_count / float(student_count) * 100
print(passing_reading_percentage)
```

```python
#calculate the percentage of students that passed math and reading
passing_math_and_reading_count = complete_school_data[(complete_school_data["math_score"] >= 70) & (complete_school_data["reading_score"] >= 70)].count()["student_name"]
overall_passing_rate = passing_math_and_reading_count / float(student_count) * 100
print(overall_passing_rate)
```

```python
#create a summary table to show the metrics in a DataFrame
summary_table = pd.DataFrame(
    {
    "Total Schools": [school_count],
    "Total Students": [student_count],
    "Total Budget": [total_budget],
    "Average Math Score": [average_math_score],
    "Average Reading Score": [average_reading_score],
    "% Passing Math": [passing_math_percentage],
    "% Passing Reading": [passing_reading_percentage],
    "% Overall Passing": [overall_passing_rate]
    }
)

#formatting of the values
summary_table["Total Students"] = summary_table["Total Students"].map("{:,}".format)
summary_table["Total Budget"] = summary_table["Total Budget"].map("{:,.2f}".format)

#displaying the table (if you do print() it doesn't display it as nicely)
summary_table
```

### School Summary

```python
#select the type per school from school_data
school_types = school_data.set_index(["school_name"])["type"]
school_types
```

```python
#calculate the count of total students per school
per_school_counts = complete_school_data["school_name"].value_counts()
per_school_counts
```

```python
#calculate the total school budget and per capita spending for each school
per_school_budget = complete_school_data.groupby(["school_name"])["budget"].mean()
per_school_capita = per_school_budget / per_school_counts
per_school_capita
```

```python
#calculate the averge test scores per school
per_school_math = complete_school_data.groupby(["school_name"])["math_score"].mean()
per_school_reading = complete_school_data.groupby(["school_name"])["reading_score"].mean()

#use print() here because there are two things you want to display
print(per_school_math)
print(per_school_reading)
```

```python
#calculate the number of students per school with math scores higher than 70%
students_passing_math = complete_school_data[(complete_school_data["math_score"]>= 70)]
school_students_passing_math = students_passing_math.groupby(["school_name"]).size()
print(school_students_passing_math)
```

```python
#calculate the number of students per school with reading scores higher than  70%
students_passing_reading = complete_school_data[(complete_school_data["reading_score"]>= 70)]
school_students_passing_reading = students_passing_reading.groupby(["school_name"]).size()
print(school_students_passing_reading)
```

```python
#calculate the number of students passing both math and reading with scores higher than 70%
students_passing_math_and_reading = complete_school_data[(complete_school_data["reading_score"] >= 70) & (complete_school_data["math_score"] >= 70)]
school_students_passing_math_and_reading = students_passing_math_and_reading.groupby(["school_name"]).size()
print(school_students_passing_math_and_reading)
```

```python
#calculate the percentages of students passing math and also studnets passing reading
per_school_passing_math = school_students_passing_math / per_school_counts * 100
per_school_passing_reading = school_students_passing_reading / per_school_counts * 100
overall_passing_rate = school_students_passing_math_and_reading / per_school_counts * 100
print(overall_passing_rate)
```

```python
#create a dataframe for all the per_school calculations with columns
per_school_summary = pd.DataFrame(
    {
        "School Types": school_types,
        "Total Students": per_school_counts,
        "School Budget": per_school_budget,
        "Budget Per Student": per_school_capita,
        "Average Math Score": per_school_math,
        "Average Reading Score": per_school_reading,
        "% of Students Passing Math": per_school_passing_math,
        "% of Students Passing Reading": per_school_passing_reading,
        "Overall Passing Rate": overall_passing_rate
    }
)

#formatting for school budget and per student budget
per_school_summary["School Budget"] = per_school_summary["School Budget"].map("${:,.2f}".format)
per_school_summary["Budget Per Student"] = per_school_summary["Budget Per Student"].map("${:,.2f}".format)

per_school_summary.index.name = None

#display the dataframe
per_school_summary
```

### Scores by School Spending

```python
#establish bins and label them
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-$630", "$630-$645", "$645-$680"]
```

```python
#create a copy of the school summary specifically for spending
school_spending_df = per_school_summary.copy()
```

```python
#use cut on the per school capita series
school_spending_df["Spending Ranges (Per Student)"] = pd.cut(per_school_capita, spending_bins, labels=labels, right=False)

#convert it to a string
school_spending_df["Spending Ranges (Per Student)"] = school_spending_df["Spending Ranges (Per Student)"].astype(str)
school_spending_df
```

```python
#calculate averages for the desired columns
spending_math_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Math Score"].mean()
spending_reading_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Reading Score"].mean()
spending_passing_math = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% of Students Passing Math"].mean()
spending_passing_reading = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% of Students Passing Reading"].mean()
overall_passing_spending = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Overall Passing Rate"].mean()
```

```python
#put this info into a dataframe
spending_summary = pd.DataFrame(
    {
        "Average Math Score": spending_math_scores,
        "Average Reading Score": spending_reading_scores,
        "% Passing Math": spending_passing_math,
        "% Passing Reading": spending_passing_reading,
        "% Overall Passing": overall_passing_spending
    }
)

#display results
spending_summary
```

### Scores by School Size

```python
#establish bins and label them
size_bins = [0, 1000, 2000, 5000]
labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```

```python
#create a copy of the school summary dataframe again
school_size_df = per_school_summary.copy()
```

```python
#use cut on the per school counts series
school_size_df["School Size"] = pd.cut(per_school_summary["Total Students"], size_bins, labels=labels, right=False)

#convert school size to a string
school_size_df["School Size"] = school_size_df["School Size"].astype(str)
school_size_df
```

```python
#calculate averages for the columns
size_math_scores = school_size_df.groupby(["School Size"])["Average Math Score"].mean()
size_reading_scores = school_size_df.groupby(["School Size"])["Average Reading Score"].mean()
size_passing_math = school_size_df.groupby(["School Size"])["% of Students Passing Math"].mean()
size_passing_reading = school_size_df.groupby(["School Size"])["% of Students Passing Reading"].mean()
size_overall_passing = school_size_df.groupby(["School Size"])["Overall Passing Rate"].mean()
```

```python
#create a dataframe that shows us school performance based on school size
size_summary = pd.DataFrame(
    {
        "Average Math Score": size_math_scores,
        "Average Reading Score": size_reading_scores,
        "% Passing Math": size_passing_math,
        "% Passing Reading": size_passing_reading,
        "% Overall Passing": size_overall_passing
    }
)
size_summary
```
